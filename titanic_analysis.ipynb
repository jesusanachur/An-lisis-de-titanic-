```python
df["first_name"] = df["Name"].str.split(",").str.get(1).str.split(".").str.get(1).str.strip()
df.head()
```

```python
df['Survived_str'] = df['Survived'].map({0: 'Murio', 1: 'Vivo'})
df[['Survived_str','title']].groupby(by=['title','Survived_str']).agg(
    numero=('title', 'count'),
)
```

```python
df['Survived_str'] = df['Survived'].map({0: 'Murio', 1: 'Vivo'})
```

```python
df[['Survived_str','Sex']].groupby(by=['Sex','Survived_str']).agg(
    numero=('Sex', 'count'),
)
```

```python
df[['Survived_str', 'Pclass', 'Sex']].groupby(by=['Sex', 'Pclass', 'Survived_str']).agg(
    numero=('Sex', 'count'),
)
```

```python
# Calculate the percentage of each group (Sex, Pclass, Survived_str) within each Pclass
grouped = df.groupby(['Sex', 'Pclass', 'Survived_str']).size().reset_index(name='count')
total_per_pclass = df['Pclass'].value_counts().sort_index()
grouped['percentage'] = grouped.apply(lambda row: row['count'] / total_per_pclass[row['Pclass']] * 100, axis=1)
grouped
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Survived', data=df)
plt.title('DistribuciÃ³n de Supervivencia')
plt.xlabel('SobreviviÃ³ (0 = No, 1 = SÃ­)')
plt.ylabel('Cantidad de Pasajeros')
plt.show()

# Porcentaje de supervivencia
survival_rate = df['Survived'].mean() * 100
print(f'Porcentaje de supervivencia: {survival_rate:.2f}%')
```

```python
# GrÃ¡fico de supervivencia por gÃ©nero
plt.figure(figsize=(10, 6))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Supervivencia por GÃ©nero')
plt.xlabel('GÃ©nero')
plt.ylabel('Cantidad de Pasajeros')
plt.legend(['No SobreviviÃ³', 'SobreviviÃ³'])
plt.show()

# Tasa de supervivencia por gÃ©nero
survival_by_sex = df.groupby('Sex')['Survived'].mean() * 100
print('Tasa de supervivencia por gÃ©nero:')
display(survival_by_sex)
```

```python
# GrÃ¡fico de supervivencia por clase
plt.figure(figsize=(10, 6))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Supervivencia por Clase de Ticket')
plt.xlabel('Clase de Ticket')
plt.ylabel('Cantidad de Pasajeros')
plt.legend(['No SobreviviÃ³', 'SobreviviÃ³'])
plt.show()

# Tasa de supervivencia por clase
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print('Tasa de supervivencia por clase de ticket:')
display(survival_by_class)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df.dropna(subset=['Age']), x='Age', bins=30, kde=True)
plt.title('DistribuciÃ³n de Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()

# Boxplot de edades por supervivencia
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=df.dropna(subset=['Age']))
plt.title('DistribuciÃ³n de Edades por Supervivencia')
plt.xlabel('SobreviviÃ³ (0 = No, 1 = SÃ­)')
plt.ylabel('Edad')
plt.show()
```

```python
# GrÃ¡fico de supervivencia por puerto de embarque
plt.figure(figsize=(10, 6))
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title('Supervivencia por Puerto de Embarque')
plt.xlabel('Puerto de Embarque')
plt.ylabel('Cantidad de Pasajeros')
plt.legend(['No SobreviviÃ³', 'SobreviviÃ³'])
plt.show()

# Tasa de supervivencia por puerto
survival_by_embarked = df.groupby('Embarked')['Survived'].mean() * 100
print('Tasa de supervivencia por puerto de embarque:')
display(survival_by_embarked)
```

```python
# Crear una columna para el tamaÃ±o de la familia
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# GrÃ¡fico de supervivencia por tamaÃ±o de familia
plt.figure(figsize=(12, 6))
sns.countplot(x='FamilySize', hue='Survived', data=df)
plt.title('Supervivencia por TamaÃ±o de Familia')
plt.xlabel('TamaÃ±o de Familia')
plt.ylabel('Cantidad de Pasajeros')
plt.legend(['No SobreviviÃ³', 'SobreviviÃ³'])
plt.show()

# Tasa de supervivencia por tamaÃ±o de familia
survival_by_family = df.groupby('FamilySize')['Survived'].mean() * 100
print('Tasa de supervivencia por tamaÃ±o de familia:')
display(survival_by_family)
```
